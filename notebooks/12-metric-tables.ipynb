```python
import pandas as pd
```

```python
def load_csv(csv_path):
    df = pd.read_csv(csv_path)
    df = df.drop([col for col in df.columns if '_step' in col], axis=1) # drop training step number columns
    df = df.replace('undefined', 0.0)
    df = df.astype('float64')
    
    return df

df = load_csv('../reports/thesis/tables/top_10_test_rouge_l_f1.csv')
df.head()
```

```python
def get_best_score(df, key):
    best_row = df.loc[df[key].idxmax()][['epoch', key]]
    return pd.DataFrame(data={'id': [key], 'epoch': [best_row['epoch']], 'score': [best_row[key]]})

get_best_score(df, 'rs_1_10 - test_rouge_L_f1')
```

```python
def get_best_scores(df, metric='test_rouge_L_f1'):
    return pd.concat([
        get_best_score(df, column)
        for column in df.columns
        if metric in column
    ]).sort_values(by='score', ascending=False).set_index('id').astype({'epoch': 'int64'})

get_best_scores(df)
```
